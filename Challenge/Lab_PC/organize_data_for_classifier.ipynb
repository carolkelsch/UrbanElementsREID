```python
import os
import shutil
```

```python
source_folder = '/media/DiscoLocal/IPCV/Carolina/unified/'

folders = ['image_train', 'image_test', 'image_query']

classes = {'Containers': 1, 'rubishbinss': 3, 'crosswalks': 2}
print(list(classes.keys()))
print(classes['Containers'])
```

```python
dest_folder = '/media/DiscoLocal/IPCV/Carolina/classification/'
os.makedirs(dest_folder, exist_ok = True)
```

```python
for f in folders:
    os.makedirs(dest_folder + f, exist_ok=True)
    for c in classes.keys():
        os.makedirs(os.path.join(dest_folder, f, c), exist_ok=True)
```

```python
for f in folders:
    src = os.path.join(source_folder, f)
    files = os.listdir(src)
    print(f'copying files for {f}')
    for c in classes.keys():
        dest = os.path.join(dest_folder, f, c)
        for file in files:
            if file.startswith(str(classes[c])):
                print(f'copying {file} to {c} folder')
                shutil.copyfile(os.path.join(src, file), os.path.join(dest, file))
```
